```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "D:\\Netflix\\NetFlix.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["director"]=df["director"].fillna("Unknown")
df["cast"]=df["cast"].fillna("Not Available")
df["country"]=df["country"].fillna("Unknown")
```

```python
df=df.dropna(subset=["date_added"])
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
df.isnull().sum()
```

```python
df["date_added"]=pd.to_datetime(df["date_added"])
df["year_added"]=df["date_added"].dt.year
df["date_added"].head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="year_added", palette="viridis", hue="type")
```

```python
# Map ratings to numerical values for averaging
rating_map = {
	'TV-Y': 0, 'TV-Y7': 1, 'G': 2, 'TV-G': 2, 'PG': 3, 'TV-PG': 3,
	'PG-13': 4, 'TV-14': 5, 'R': 6, 'TV-MA': 7, 'NC-17': 8
}
df['rating_num'] = df['rating'].map(rating_map)

avg_rating = df.groupby("year_added")["rating_num"].mean().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data=avg_rating, x="year_added", y="rating_num", marker="o")
```
